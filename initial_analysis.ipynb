```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pn
# Display plots inline
%matplotlib inline
```

# Metadata Setup

```python
from src.data import  DATA_PATH

#download_data(force_download=False)


movie_columns = ["WikipediaId", "FreebaseId", "MovieName", "ReleaseDate",
                 "Revenue", "Runtime", "Languages", "Countries", "Genres"]
character_columns = ["WikipediaId", "FreebaseId", "ReleaseDate", "CharacterName",
                     "DateOfBirth", "Gender", "Height", "Ethnicity", "ActorName",
                     "AgeAtRelease", "CharacterActorMapId", "CharacterId", "ActorId"]
plot_columns = ["WikipediaId", "PlotSummary"]


movie_metadata = pd.read_csv(DATA_PATH / "MovieSummaries" / "movie.metadata.tsv", sep="\t", names=movie_columns )
character_metadata = pd.read_csv(DATA_PATH / "MovieSummaries" / "character.metadata.tsv", sep="\t", names=character_columns)
plot_summaries = pd.read_csv(DATA_PATH / "MovieSummaries" / "plot_summaries.txt", sep="\t", names=plot_columns)


print("Movie Metadata:")
display(movie_metadata.head())

print("\nCharacter Metadata:")
display(character_metadata.head())

print("\nPlot Summaries:")
display(plot_summaries.head())


print("\nMovie Metadata Info:")
movie_metadata.info()

print("\nCharacter Metadata Info:")
character_metadata.info()

print("\nPlot Summaries Info:")
plot_summaries.info()

```

# Data Cleaning

WIP: perform data cleanup after the selected analysis

```python
def parse_nested_column(column_data):
    parsed_data = []
    for entry in str(column_data).split(', '):
        try:
            parsed_data.append(entry.split(':')[1])
        except IndexError:
            continue
    return parsed_data

movie_missing = movie_metadata.isnull().sum()
character_missing = character_metadata.isnull().sum()

print("Missing Values in Movie Metadata:")
print(movie_missing)

print("\nMissing Values in Character Metadata:")
print(character_missing)

print("\nMissing Values in Plot Summaries:")
print(plot_summaries.isnull().sum())


movie_metadata['Runtime'].fillna(movie_metadata['Runtime'].mean(), inplace=True)
character_metadata['Height'].fillna(character_metadata['Height'].median(), inplace=True)
movie_metadata['ReleaseDate'] = pd.to_datetime(movie_metadata['ReleaseDate'], errors='coerce')
character_metadata['DateOfBirth'] = pd.to_datetime(character_metadata['DateOfBirth'], errors='coerce')




movie_missing_df = pd.DataFrame({
    'Column': movie_missing.index,
    'MissingValues': movie_missing.values
}).sort_values(by='MissingValues', ascending=False)

character_missing_df = pd.DataFrame({
    'Column': character_missing.index,
    'MissingValues': character_missing.values
}).sort_values(by='MissingValues', ascending=False)


sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=movie_missing_df, x='Column', y='MissingValues', color='skyblue')
plt.title("Missing Values in Movie Metadata")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Missing Values")
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
sns.barplot(data=character_missing_df, x='Column', y='MissingValues', color='salmon')
plt.title("Missing Values in Character Metadata")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Missing Values")
plt.tight_layout()
plt.show()

```

# Data Screening and Initial Analysis

```python
print("Movie Metadata Summary Statistics:")
display(movie_metadata[['Revenue', 'Runtime']].describe())

print("Character Metadata Summary Statistics:")
display(character_metadata[['Height', 'AgeAtRelease']].describe())

print("Top 10 Most Common Movie Genres:")
display(movie_metadata['Genres'].explode().value_counts().head(10))

print("\nTop 10 Most Common Languages:")
display(movie_metadata['Languages'].explode().value_counts().head(10))

print("\nGender Distribution in Characters:")
display(character_metadata['Gender'].value_counts())


sns.boxplot(data=movie_metadata, x='Revenue')
plt.title('Box Office Revenue Distribution')
plt.show()

# Visualize genre distribution
genre_counts = movie_metadata['Genres'].explode().value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Top 10 Movie Genres')
plt.xlabel('Frequency')
plt.show()
```
